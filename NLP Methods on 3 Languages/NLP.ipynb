```python
# load data
filename = '/content/drive/MyDrive/data_ananna/lotr.txt'
file = open(filename, 'rt')
text = file.read()
file.close()
```

```python
# split into words
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print(tokens[10:200])
```

```python
# convert to lower case
tokens = [w.lower() for w in tokens]
print(tokens[10:200])
```

```python
# remove punctuation from each word
import string
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in tokens]
print(stripped[10:200])
```

```python
# remove remaining tokens that are not alphabetic
words = [word for word in stripped if word.isalpha()]
print(words[:200])
```

```python
# filter out stop words
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = [w for w in words if not w in stop_words]
print(words[:200])
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
def word_lemmatizer(text):
     lem_text=[lemmatizer.lemmatize(i) for i in text]
     return lem_text
```

```python
words=word_lemmatizer(words)
```

```python
print(words[:200])
```

```python
# finding the frequency distinct in the tokens
# Importing FreqDist library from nltk and passing token into FreqDist
from nltk.probability import FreqDist
fdist = FreqDist(words)
fdist
```

```python
# To find the frequency of top 30 words
fdist1 = fdist.most_common(30)
fdist1
```

```python
for token in words[:100]:
  print(nltk.pos_tag([token]))
```

```python
for token in words[:100]:
  pos_token=nltk.pos_tag([token])
  nn_tagged = [(word,tag) for word, tag in pos_token if tag.startswith('NN')]
  print(nn_tagged)
```

```python
for token in words[:100]:
  pos_token=nltk.pos_tag([token])
  nn_tagged = [(word,tag) for word, tag in pos_token if tag.startswith('VB')]
  print(nn_tagged)
```

NLP on Hindi Language Corpus

```python
# load data
filename = '/content/drive/MyDrive/data_ananna/hindi.txt'
file = open(filename, 'rt')
text = file.read()
file.close()
```

```python
# split into words
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print(tokens[3:200])
```

```python
# remove punctuation from each word
import string
table = str.maketrans('', '', string.punctuation)
words = [w.translate(table) for w in tokens]
print(words[:100])
```

```python
# finding the frequency distinct in the tokens
# Importing FreqDist library from nltk and passing token into FreqDist
from nltk.probability import FreqDist
fdist = FreqDist(words)
fdist
```

```python
# To find the frequency of top 30 words
fdist1 = fdist.most_common(30)
fdist1
```

```python
import nltk
nltk.download('indian')
from nltk.tag import tnt
from nltk.corpus import indian
train_data = indian.tagged_sents('hindi.pos')
tnt_pos_tagger = tnt.TnT()
tnt_pos_tagger.train(train_data)
```

```python
tagged_words = (tnt_pos_tagger.tag(words[:200]))
nn_tagged = [(word,tag) for word, tag in tagged_words if tag.startswith('NNP')]
print(tagged_words)
```

```python
# load data
filename = '/content/drive/MyDrive/data_ananna/অঁহক – হুমায়ূন আহমেদ.txt'
file = open(filename, 'rt')
text = file.read()
file.close()
```

```python
from bnlp import NLTKTokenizer
from bnlp.corpus import stopwords
from bnlp.corpus.util import remove_stopwords

bnltk = NLTKTokenizer()

remove_sw = remove_stopwords(text, stopwords)

word_tokens = bnltk.word_tokenize(text)
sentence_tokens = bnltk.sentence_tokenize(text)
print(sentence_tokens)

```

```python
print(word_tokens[:200])
```

```python
print(remove_sw[:200])
```

```python
# remove punctuation from each word
import string
table = str.maketrans('', '', string.punctuation)
words = [w.translate(table) for w in remove_sw]
print(words[:200])
```

```python
# finding the frequency distinct in the tokens
# Importing FreqDist library from nltk and passing token into FreqDist
from nltk.probability import FreqDist
fdist = FreqDist(words)
fdist
```

```python
# To find the frequency of top 30 words
fdist1 = fdist.most_common(30)
fdist1
```

```python
from bnlp import POS
bn_pos = POS()
model_path = "/content/drive/MyDrive/data_ananna/bn_pos.pkl"
res = bn_pos.tag(model_path, text)
nn_tagged = [(word,tag) for word, tag in res if tag.startswith('NC')]
print(res)
```
